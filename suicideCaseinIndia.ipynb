```python
# Creating a dictionary mapping states to regions
region_mapping = {
    'A & N Islands': 'Southern',
    'Andhra Pradesh': 'Southern',
    'Arunachal Pradesh': 'North Eastern',
    'Assam': 'North Eastern',
    'Bihar': 'Eastern',
    'Chandigarh': 'Northern',
    'Chhattisgarh': 'Central',
    'D & N Haveli': 'Western',
    'Daman & Diu': 'Western',
    'Delhi (Ut)': 'Northern',
    'Goa': 'Western',
    'Gujarat': 'Western', 
    'Haryana': 'Northern',
    'Himachal Pradesh': 'Northern',
    'Jammu & Kashmir': 'Northern',
    'Jharkhand': 'Eastern',
    'Karnataka': 'Southern',
    'Kerala': 'Southern',
    'Lakshadweep': 'Southern',
    'Madhya Pradesh': 'Central',
    'Maharashtra': 'Western',
    'Manipur': 'North Eastern',
    'Meghalaya': 'North Eastern',
    'Mizoram': 'North Eastern',
    'Nagaland': 'North Eastern',
    'Odisha': 'Eastern',
    'Puducherry': 'Southern',
    'Punjab': 'Northern',
    'Rajasthan': 'Western',
    'Sikkim': 'North Eastern',
    'Tamil Nadu': 'Southern',
    'Total (All India)': '',
    'Total (States)': '',
    'Total (Uts)': '',
    'Tripura': 'North Eastern',
    'Uttar Pradesh': 'Northern',
    'Uttarakhand': 'Northern',
    'West Bengal': 'Eastern'
}
```

```python
# Creating a function that maps states to region
def map_state_to_region(state):
    return region_mapping.get(state, 'Unknown')

df['Region'] = df['State'].apply(map_state_to_region)
```

```python
edu_df = df[df['Type'].isin(['Diploma', 'Graduate', 'Hr. Secondary/Intermediate/Pre-Universit', 'Middle', 'No Education', 'Primary'])]
# edu_df

education = edu_df['Type'].unique()
males = []
females = []

for e in education:
    males.append(edu_df[(edu_df['Type'] == e) & (edu_df['Gender'] == 'Male')]['Total'].sum())
    females.append(edu_df[(edu_df['Type'] == e) & (edu_df['Gender'] == 'Female')]['Total'].sum())


plt.figure(figsize=(10, 6))
plt.bar(education, males, label='Male', color='red', edgecolor='black')
plt.bar(education, females, bottom=males, label='Female', color='orange', edgecolor='black')
plt.title('Total no of cases based on Education by Gender')
plt.xlabel('Education')
plt.ylabel('Total')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.show()
```

```python
eco_df = df[df['Type'].isin(['Poverty', 'Unemployment', 'Bankruptcy or Sudden change in Economic'])]
# eco_df

economic = eco_df['Type'].unique()
males = []
females = []

for e in economic:
    males.append(eco_df[(eco_df['Type'] == e) & (eco_df['Gender'] == 'Male')]['Total'].sum())
    females.append(eco_df[(eco_df['Type'] == e) & (eco_df['Gender'] == 'Female')]['Total'].sum())


plt.figure(figsize=(10, 6))
plt.bar(economic, males, label='Male', color='red', edgecolor='black')
plt.bar(economic, females, bottom=males, label='Female', color='orange', edgecolor='black')
plt.title('Total no of cases based on Economic and Employment Issues by Gender')
plt.xlabel('Economic & Employment Issues')
plt.ylabel('Total')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.show()
```
